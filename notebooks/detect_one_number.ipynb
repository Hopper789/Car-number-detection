```python
from ultralytics import YOLO
from glob import glob
from numpy import random
import cv2
from matplotlib import pyplot as plt

model = YOLO('../weights/best.pt')

# –ü—É—Ç—å –∫ —Ç–µ—Å—Ç–æ–≤—ã–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º
image_files = glob('../data/datasets/andrewmvd/car-plate-detection/versions/1/images/*.png')
subset = random.choice(image_files, size=5)
print(f"\n--- üß™ –ó–∞–ø—É—Å–∫–∞–µ–º –ø—Ä–æ–≤–µ—Ä–∫—É –Ω–∞ {len(subset)} —Å–ª—É—á–∞–π–Ω—ã—Ö —Ç–µ—Å—Ç–æ–≤—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è—Ö ---")
for img_path in subset:
    # –ó–∞–ø—É—Å–∫–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Å –ø–æ—Ä–æ–≥–æ–º —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ 0.5
    img = cv2.imread(img_path)
    preds = model.predict(img, conf=0.5, verbose=False) 
    annotated_frame = preds[0].plot()

    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
